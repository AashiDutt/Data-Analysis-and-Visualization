**About the dataset:**

The market research team at AdRight is assigned the task to identify 
the profile of the typical customer for each treadmill product offered by CardioGood Fitness. The market research team decides to investigate whether there are differences across the product lines with respect to customer characteristics. The team decides to collect data on individuals who purchased a treadmill at a CardioGoodFitness retail store during the prior three months.

 The data are stored in the CardioGoodFitness.csv file. The team identifies the following customer variables to study: 
 
 


1.   product purchased, TM195, TM498, or TM798;
2.   gender; age, in years;
3. education, in years;
4. relationship status, single or partnered;
5. annual household income ($);
6. average number of times the customer plans to use the treadmill each week; 
7. average number of miles the customer expects to walk/run each week;  
8. self-rated fitness on an 1-to-5 scale, where 1 is poor shape and 5 is excellent shape.

```
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```
sns.countplot(x = 'Product', hue = 'Gender', data = df)
```

```
sns.pairplot(df)
```

```
# correlation of all columns

df.corr()
```

```
plt.figure(figsize=(12,6))
```

```
# bool or rectangular dataset,
sns.heatmap(df.corr(), annot = True)
```

```
sns.scatterplot(x ='Age', y ='Income', hue ='Product', data = df)
```

```
#Univariate Analysis
#categorical variables 
plt.figure(figsize=(14,7))
df['Product'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(8,8))
plt.title("Pie chart of Product Sales")
plt.show()
```

Max sales for TM195

```
plt.figure(figsize = (7,7))
sns.countplot(df['Gender'], hue=df["MaritalStatus"]).set(title= "Marital status by Gender")
```

```
plt.figure(figsize = None)
sns.catplot(x ='Usage', y = 'Income', col = 'Gender',hue = 'Product', kind = "bar", data = df)
```

```
#scatter plot between income, age ,product and usage
sns.relplot(x="Age", y="Income", hue="Product", size="Usage",
            sizes=(40, 400), alpha=.5,
            height=6, data=df).set(title='INCOME BY AGE ,PRODUCT AND USAGE');
```

Products TM498 and TM195 were brought by people having income between 30k to 70 k.

While product TM798 was brought by people having income above 50k.


Thus from all the charts we can say that product TM195 had maximum sales because it was one of the affordable product on the list.
