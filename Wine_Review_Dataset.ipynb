**Pandas work on row first , column second concept for indexing.**

#Accessor Operators in Pandas ( loc and iloc)

# 1. Index Based Selection (iloc) - selecting data based on its numerical position in dataset.

```python


reviews.iloc[0] # gives all the info about the data present in 1st row of the dataset.

```

```python
reviews.iloc[:,0]   # all rows of 1st column
```

```python
reviews.iloc[1:3, 0]  # 2nd and 3rd rows  
```

```python
reviews.iloc[-5:] # Get elements from end of dataset
```

```python
reviews.iloc[[0,1,2],0]  # pass a list of rows you need from 1st column
```

# 2. Label based selection(loc) - uses data as index

```python
reviews.loc[0, 'country'] # get 1st entry in country column
```

```python
reviews.loc[[0,1,2], 'country']  # pass list of indices for which you need value from country column
```

```python
reviews.loc[:, ['country','points','variety']] # returns all rows for the named columns
```

**Difference between iloc and loc**

**iloc** 



1.   if index = [0:10], returns values of entries 0 to 9.
2.   Can take only integers for indexing



**loc** 

1.   if index =[0:10], returns values of entries from 0 to 10.
2.   takes both integer as well as non- integer values.







```python
reviews.loc[0:5] # returns all rows from index 0 to 5.
```

```python
reviews.iloc[0:5] # returns all rows from index 0 to 4.
```

**Manipulating Index**

```python
reviews.set_index("country") # set any column as index
```

```python
# Conditional Selection 
reviews.country == 'Italy'  # returns True or False for each element that satisfies the condition.
```

```python
reviews.loc[reviews.country == "Italy"] # above code can be used in this way as well.
```

```python
# And and OR  operator

reviews.loc[(reviews.country == 'Italy') & (reviews.points >= 90)]
```

```python
reviews.loc[(reviews.country == 'Italy') | (reviews.points >= 90)]
```

```python
# Built-In Conditional Selectors
# 1. isin - lets you select data whose value is in list of values 

reviews.loc[reviews.country.isin(['Italy', 'France'])]
```

```python
# 2. isnull and notnull

reviews.loc[reviews.region_2.isnull()]
```

```python
reviews.loc[reviews.price.notnull()]
```

```python
# Summary functions and Maps

reviews.country.describe()
```

```python
reviews.points.mean()
```

```python
# list of unique value

reviews.points.unique()
```

```python
# count the values 

reviews.country.value_counts()
```

```python
# Maps - Returns Series

reviews_points_mean = reviews.points.mean()
reviews.points.map(lambda points: points - reviews_points_mean)
```

```python
# Apply() operator - returns DataFrame by transformation

def remean_points(row):
  row.points = row.points - reviews_points_mean
  return row
reviews.apply(remean_points, axis = 'columns')
```

```python
# Grouping and Sorting

# 1. Groupby() 

reviews.groupby('points').points.count()
```

```python
# group points by min price

reviews.groupby('points').price.min()
```

```python
# Manipulating groupby using apply() method

reviews.groupby('winery').apply(lambda df:df.country.iloc[0])
```

```python
reviews.groupby(['country','province']).apply(lambda df:df.loc[df.points.idxmax()])
```

```python
# Agg() run a bunch of functions

reviews.groupby(['country']).price.agg([len, min, max])
```

```python
# Multi Indexes

x = reviews.groupby(['country','province']).description.agg([len])

mi = x.index
type(mi)
```
